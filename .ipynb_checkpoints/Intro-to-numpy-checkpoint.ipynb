### Exercise 8
#### Random vector and plots
a. Create a random column vector with dimension 100, according to a Normal
distribution, with zero mean and standard deviation 1. Plot the result. Experiment
with different colors, line styles, and markers

```python
N = 100
x = np.random.randn(N,1) # N = rader, 1 = kolonner

sigma = 3 # Standard deviation
mu = 20 # Mean

x = sigma * x + mu
```

```python
import matplotlib.pyplot as plt

plt.plot(x,'r')
plt.xlabel('sample')
plt.ylabel('x')
plt.legend('x')
plt.grid()
```

b. Create another random column vector with dimension 100, according to a uniform
distribution, mean 5 and standard deviation 2. Plot the result.

```python
y = np.random.randn(N,1)
sigma2 = 2
mu2 = 5
y = sigma2*y+mu2
```

```python
plt.plot(y,'--b')
plt.xlabel('sample')
plt.ylabel('y')
plt.legend('y')
plt.grid()
```

c. Plot both vectors in the same figure and with different colors/markers. Use legend
command to label each of the plots. 

```python
plt.plot(y,'b',label='y')
plt.plot(x,'r',label='x')
plt.xlabel('sample')
plt.ylabel('x and y')
plt.grid()
plt.legend() # aktiverer labels som nevn ovenfor
```

### Exercise 9
#### Plot the signal
ùë¶ = 2 sin(ùúîùë° + 1)

Between time ùë° = 0ùë† and ùë° = 10ùë†, with ùúî = 2.

Hint: Decide on how many samples you need. Experiment what happens when changing the
values of w. One option is to use a for loop to create the signal but it is much simpler and
efficient to make an operation on the time vector t. 

```python
w = 2
t = np.linspace(0,10,N) # intervall fra 0 til 10s, med 100 steg
y = 2*np.sin(w*t+1)

plt.plot(t,y, label='2sin(wt+1)')
plt.ylabel('y(t)')
plt.xlabel('t [s]')
plt.legend()
plt.grid()
```

Example on how to plot several plots in one figure with random w.

```python
N = 1000 # number of samples,
t = np.linspace(0,10,N)
Nplots = 5
plt.figure()
for i in range(Nplots):
    w = np.random.random()
    y = 2*np.sin(w*t+1)
    plt.plot(t,y,'-',label="w={0:2.2}".format(w)) 
plt.grid()
plt.xlabel('t')
plt.ylabel('y[t]')
plt.legend()
```

### Exercise 10
#### Plot the signal
$ùë¶ = 2ùë•^3‚àí ùë•^2 + ùë• ‚àí 1$

In the interval ùë• ‚àà [‚àí20,20]. 

Hint: Writing a dot after a vector means the operation applies
to the vector elements. For instance $x**2$ takes the square of all elements in vector x; and $x**3$ its cubic exponent. Also if x and y are two vectors of the same dimensions,$ x*y‚Äù indicates the element-wise product of the elements in vectors x and y.

```python
N = 1000 #samples
x = np.linspace(-20,20,N) #samplevector
y = 2*x**3 - x**2 + x - 1

plt.plot(x,y,'--y',label='y(x)')
plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('y(x)')
```
